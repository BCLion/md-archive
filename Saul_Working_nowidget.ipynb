````python
from unsloth import FastLanguageModel
from peft import PeftModel
import torch


# Load the trained checkpoint
max_seq_length = 2048
load_in_4bit = True
dtype = None # Define dtype here, following the pattern from the first model loading cell

model, tokenizer = FastLanguageModel.from_pretrained(
    "Shogun9000/saul-7b-bc-family-law-v2",
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=load_in_4bit,
)

# Your test prompt (paste your 2023bcpc26 snippet here)
prompt =  """### SYSTEM ROLE
You are an elite Senior Counsel with 35+ years in Canadian Family Law (BC focus). You act as a "Teacher" generating extremely high-fidelity training data for a student model (Saul-7B).

### CORE MISSION
Produce the highest quality, most legally precise analysis possible. Strike fear into those who weaponize the family court system while protecting vulnerable parents and children.

### ANALYSIS PROTOCOL â€” MUST FOLLOW
1. Read the full case markdown carefully.
2. Perform deep silent analysis first (jurisdiction, best interests test s.37 FLA, violence definitions, etc.).
3. In the counsels_analysis <think> block, explicitly label each component as follows:
   "Issue:"
   "Test:" (the statutory/precedential/legal test or requirements that must be met)
   "Theory:" (the strategic legal narrative/doctrine that explains why your side wins when the test is applied, e.g., "Theory: Post-Barendregt child-centric analysis prioritizes support network and stability over continuity when...")
   "Application:"
   "Conclusion:"
4. Output **exactly** in this order â€” nothing else before, between, or after:


### LEGAL REASONING TRACE
[Here write 400â€“800 words of pure, ruthless legal reasoning:
â€¢ JURISDICTIONAL SCRUTINY â€” BCPC vs BCSC differences in depth & precedent weight
â€¢ STATUTORY SUBSUMPTION â€” map facts â†’ specific sections (FLA s.37, s.46, s.182â€“185, etc.)
â€¢ PRECEDENTIAL ALIGNMENT â€” how judge handled Gordon/Barendregt/Young/etc. â†’ correct application or gap?
â€¢ ARGUMENT STRENGTH â€” why winning side prevailed (doctrine + evidence)
â€¢ JUDICIAL SENTIMENT â€” exact quotes showing disdain/frustration/deference
â€¢ IRAC/CoT SYNTHESIS â€” full Issue-Rule-Application-Conclusion walkthrough
]

Then, immediately after (new line, no extra text):

{raw JSON string following the schema below â€” no ```json, no explanation}

### SCHEMA (must follow exactly):
{
    # --- IDENTITY ---
    "case_id": "Unique ID slug (e.g., 2023_BCPC_167)",
    "case_name": "Full title (e.g., A.P. v. C.N.)",
    "citation": "Neutral citation (e.g., 2023 BCPC 167)",
    "source_court": "Strictly one of: 'BCPC', 'BCSC', 'BCCA', 'SCC'",

    # --- CONTEXT ---
    "judgment_date": "Date of decision (YYYY-MM-DD or readable format)",
    "judge_clean": "Judge name (cleaned)",
    "child_age_approx": "Strictly one of: '<10', '10-11', '12-13', '14+', 'multiple', 'unknown'",
    "child_demographics": "Free-text: number of children, genders, any special needs notes",
    "special_needs_flag": "Y/N + type (e.g., 'Y - autism high-functioning')",
    "relocation_flag": "Strictly 'Y' or 'N'",
    "applicant_role": "Strictly 'mother', 'father', or 'joint'",

    # --- THE "WHAT" (LEGAL ISSUES) ---
    "legal_issues_identified": "FRAME LIKE A LAWYER: Instead of 'moving house', use 'Unauthorized Relocation (s. 65 FLA)'. Instead of 'money', use 'Conduct Orders (s. 226 FLA)'. Identify the legal nature of the dispute.",
    "primary_outcome": "Strictly one of: 'granted', 'dismissed', 'varied', 'partial'",
    "outcome_for_applicant": "Strictly one of: 'win', 'lose', 'partial'",
    "summary_holding": "1-2 sentence key principle",

    # --- THE "WHY" (LEGAL THEORY) ---
    "winning_argument_type": "ELITE LEVEL: Identify the specific LEGAL THEORY or DOCTRINE that caused the victory. Examples: 'Established pattern of non-compliance', 'Good faith relocation lacking', 'Status quo & stability override views', 'Maximum contact principle'. Cite the reasoning.",
    "failed_argument_type": "ELITE LEVEL: Identify the specific LEGAL DEFICIENCY that caused the loss. Examples: 'Failed to meet burden of proof', 'Lack of specific parenting plan', 'Self-representation deemed inexcusable', 'Relocation deemed unilateral and deceptive'. Focus on legal error.'",

    # --- ANALYSIS (CRITICAL FIELDS) ---
    "judicial_disdain": "CRITICAL: Identify explicit judicial criticism, strong disapproval of conduct, or tone indicating frustration. Look for 'troublesome', 'disrepute', 'frustrated'. Use direct quotes.",
    "views_weight": "Strictly one of: 'followed', 'not determinative', 'inappropriate', 'significant but overridden'",

    # --- PRECEDENT LAYER (GAP ANALYSIS) ---
    "scc_precedent_applied": "Y/N + which (e.g., 'Y - Barendregt v. Grebliunas factors', 'Y - Gordon v. Goertz')",
    "precedent_gap": "CRITICAL FOR GAP ANALYSIS: Analyze if the court followed binding SCC precedents (e.g., Gordon, Barendregt). If the court deviates (e.g., ignores child views in relocation), state the deviation. Otherwise, state 'None noted'.",

    # --- STRATEGIC TAKEAWAY ---
    "strategic_takeaway": "ELITE LAWYER TASK: Identify the *legal principle* or *lesson*. How does this impact future cases? Mention specific Acts (FLA s. 61, s. 231) or concepts. 1-2 high-level sentences.",

    # --- ELITE LAWYER ANALYSIS ---
    "counsels_analysis": (
      "ELITE LAWYER TASK: Detailed step-by-step legal reasoning using IRAC (Issue-Rule-Application-Conclusion) framework "
      "combined with explicit Chain-of-Thought. Explain 'Why' the outcome occurred using specific statutes (FLA s.X, Divorce Act s.Y), "
      "binding precedents (Gordon, Barendregt, etc.), and factual connections. Identify strengths/weaknesses in judicial reasoning. "
      "2â€“5 paragraphs, sharp, professional, no hedging."
    ),

    # --- METRICS ---
    "judge_leaning_score": "Metric (-1 to +1). -1 = Biased against Applicant, +1 = Biased for Applicant.",
    "extraction_confidence": "Strictly one of: 'High', 'Medium', 'Low'",
    "full_text_path": "URL"
}

### OUTPUT RULES â€” MUST OBEY (VIOLATE THESE AND YOU FAIL):
â€¢ LEGAL REASONING TRACE must be 400â€“800 words minimum; be exhaustive and unsparing in exposing patterns of manipulation, alienation, and judicial frustration. Use direct quotes liberally.
â€¢ Output **exactly** in this strict order â€” nothing else before, between, or after:
  1. ### LEGAL REASONING TRACE (400â€“800 words of pure, ruthless legal reasoning including JURISDICTIONAL SCRUTINY, STATUTORY SUBSUMPTION, PRECEDENTIAL ALIGNMENT, ARGUMENT STRENGTH, JUDICIAL SENTIMENT, IRAC/CoT SYNTHESIS).
  2. Exactly **one blank line** after the trace.
  3. **ONLY** the raw JSON object â€” no ```json, no backticks, no extra text, no explanation, no preamble.
â€¢ JSON must be **perfectly valid** (no trailing commas, proper escaping, no syntax errors).
â€¢ Output the raw JSON object **directly** â€” do NOT wrap it in quotes, backticks, markdown, or any other characters. It must start with { and end with }.
â€¢ In counsels_analysis, always ground every factual claim, quote, or legal point with **specific references** to the provided case text (e.g., [para 42], [p. 17], [line 238]). Do NOT cite external cases unless explicitly present in the text.
â€¢ Always wrap the **full IRAC/CoT reasoning** inside counsels_analysis with <think>...</think> tags first, then continue with the final distilled legal analysis. Example structure inside counsels_analysis:
  <think>Full step-by-step IRAC/CoT reasoning here with [para X] citations...</think>

  Final analysis: [distilled legal conclusion, sharp and professional]




Now analyze this case:

IMPORTANT: Do NOT repeat or copy the input text. Do NOT continue the case narrative.
Start your analysis immediately with <think> containing full IRAC/CoT reasoning, including explicit "Issue:", "Theory:", "Rule:", "Application:", "Conclusion:" labels. Ground everything with [para X] citations. Then provide the distilled final analysis.

Now analyze this case:
Citation:

â˜¼C.C. v. C.H.C.

2025 BCPC 9

Date:

â˜¼20250121

File No:

[Omitted for Publication]

Registry:

[Omitted for Publication]

## IN THE PROVINCIAL COURT OF BRITISH COLUMBIA

## IN THE MATTER OF

THE FAMILY LAW ACT , S.B.C. 2011 c. 25

BETWEEN:

C.C.

APPLICANT

AND:

C.H.C.

RESPONDENT

## REASONS FOR JUDGMENT OF THE

## HONOURABLE JUDGE D. REEVES

Counsel for the Applicant:

Jean-Francois Normand Maxwell

Counsel for the Respondent:

Bryan M. Crampton

Counsel for the Children:

D. Vance Grimard

Places of Hearing:

[Omitted for Publication] and Fort St John

(by way of MS Teams), B.C.

Dates of Hearing:

July 10-13, October 3, 2023,

January 18, February 16, March 12,

May 13-17, October 1,2, 2024

Date of Judgment:

January 21, 2025

- [1] C.C. and C.H.C. are the parents of P.S.C.C., born [omitted for publication], and T.G.C.C., born [omitted for publication]. The parties met in 1999 while living in the Lower Mainland, entered a marriage like relationship in September 2002 and married in 2004. In 2021, C.C. made allegations of an extra-marital affair against C.H.C. and in April of that year they began discussing separation and divorce, and shortly after chose to live separated but within the same home.
- [2] In January of 2022, C.H.C. began severing financial aspects of their union due to a fear that C.C. could inappropriately access back pay he was to receive from a recently resolved workplace pay issue. C.H.C. testified that early in 2022, shortly after C.C. quoted 'unreasonable' support figures after speaking to legal counsel, conversations between them broke down. To describe their relationship following this as acrimonious and dysfunctional would be a significant understatement and does not begin to approach the visible hostility they shared thereafter against each other.
- [3] In March 2022, C.C. began surreptitiously recording family conversations to document what she described as a shift in the home dynamic against her, and entered 17 recordings into evidence. C.C. also accused C.H.C. of potentially criminal and workplace related acts for which he was investigated but no charges were laid and no job action was taken. Following a particularly high conflict argument among all the family, videotaped by T.G.C.C. and entered into evidence, C.H.C. and the adolescent children moved out of the shared home in late August 2022.
- [4] As of the date of this decision, two and a half years later, P.S.C.C. and T.G.C.C. have not initiated any contact with their mother, have refused all efforts by her to contact them, and have been permitted by C.H.C. not to attend court ordered counselling. C.H.C. has disregarded or self-interpreted a court order requiring counselling, refused to advise C.C. of significant, or any, events affecting the children, and has maintained a steadfast barrier between her and the adolescent children ostensibly at the children's urging.
- [5] Caught up in all of this are the two teenage children who have attempted to navigate their family situation as best they could, but given an observable lack of mature

guidance or support by both parents they find themselves in a 'no-mans' land of family conflict. Throughout this proceeding, these young adults have been represented by counsel appointed from the Child and Youth Legal Centre ('CYLC'), a non-profit organization that provides pro-bono legal services to BC children and youth. Both young adults have stated they wish no contact of any nature with their mother, repeatedly through counsel and in a judicial interview. As of the date of this decision, P.S.C.C. is now 19 years old and thus is no longer a child of the marriage as defined in the Family Law Act , and T.G.C.C. is now 16.

[6] C.C. asks this court to find C.H.C. has engaged in parental alienation and seeks a court order directing that T.G.C.C. live with her and be compelled to participate in a Family Bridges program to reconnect them, along with child and spousal support. C.H.C. instead suggests the adolescent children's decision to sever all contact with their mother is actually realistic estrangement brought about by C.C. through her unsubstantiated allegations against him, controlling and obstinate nature, and abuse of his and the children's confidence by surreptitiously recording private communications within the family home. C.H.C. seeks that spousal support be set at the lower end, the children continue to live with him, and child support flow accordingly. He also seeks additional income to be imputed to C.C., rejecting her claim of reduced employment due to medical reasons.

- [7] Allegations of family violence have been made by both parties against the other, which must also be considered by this court. In coming to my decision I will examine matters in the following order:
- a.  Background of the Parties and their Relationship;
- b.  Allegations of Family Violence;
- c.  Views of the Adolescent Children;
- d.  Estrangement vs. Alienation;
- e.  Parenting Time and Family Bridges Counselling;
- f. Imputation of Income, Child and Spousal Support; and
- g.  Decision.

## A. Background of the Parties and their Relationship

- [8] Both C.H.C. and C.C. testified to a normal family situation prior to their decision to separate. Shortly after they first met, C.H.C. enrolled in his current job and C.C. commenced part-time work at that same employer. The couple moved to Coquitlam. B.C., for C.H.C.'s job, and C.C. began working at area locations on an on-call basis. While the children were young C.H.C. typically worked a 12-hour, four-days-on / fourdays-off schedule with two days and two nights worked, or several days in a row followed by a few days off while he was in another role. Both agreed that C.C. was primarily a stay-at-home mom caring for the children until they commenced school.
- [9] The family moved to their current location late in 2013, for C.H.C.'s career, where he was subsequently promoted by his employer. C.C. began working part time in local retail stores and would travel to the lower mainland to work periodic shifts for the same employer in order to maintain her security clearance. In 2018, C.C. began regular, part time work with the same employer and in the same building as C.H.C. Both parties agreed that C.H.C. took on more parental duties alongside C.C. including grocery shopping and meal preparation as the children grew older. At no time did C.H.C. have concerns for C.C.'s parenting of the children or their safety while in her care.

## B. Allegations of Family Violence

- [10] Family violence is defined at s.1 of the Family Law Act , SBC 2011 c25, and of import given the allegations and evidence I have heard include, with or without an intent to harm a family member, (a) physical abuse, or (c) attempts to physically abuse a family member, (d) psychological or emotional abuse of a family member, including (i) intimidation, harassment, coercion or threats, and (ii)unreasonable restrictions on, or prevention of, a family member's financial or personal autonomy. Other types of conduct that courts have recognized as a form of family violence include the concept of a pattern of coercive and controlling behaviour such as a) Making numerous unsubstantiated allegations against the other party; b) Unilaterally changing court-ordered parenting time terms without justification; and, c) Regularly engaging in behaviour that has the effect of undermining the other parent's authority or influence and alienating the child from that

parent ( Piaskoski v. Piaskoski , 2024 ONSC 5474 (CanLII) at paras 23 - 24, citing with approval M.A.B. v. M.G.C. , 2022 ONSC 7207, at para. 184).

- [11] Where allegations of family violence are made, section 38 of the FLA requires that in considering the best interests of the child under s.37(2)(g) and (h), I must consider (a) the nature and seriousness of the family violence; (b) the recency of the family violence; (c) its frequency, if any psychological or emotional abuse constitutes or is evidence of a pattern of coercive and controlling behaviour directed at a family member; (f) if the child was exposed to family violence directed at others; (g) the resulting harm to the child's physical, psychological and emotional safety, security and well-being; (h) steps the person responsible for family violence has taken to prevent further family violence from occurring; and (i) other relevant matters.
- [12] C.C.'s evidence described an overall positive relationship with both of the children as they grew up, taking them to doctor appointments, extracurricular activities and play dates. This characterization was not challenged by C.H.C.'s evidence. This was in contrast to evidence heard from the adolescent children themselves in the secret recordings made by C.C., or provided to the court on their behalf through CYLC legal counsel summaries of their positions.
- [13] C.C. testified that after January 2022, communication within the family, and in particular between her and the two adolescent children, began to break down which she attributed to C.H.C. C.C. suggested that C.H.C. would passively undermine her to the children with dismissive comments and body language, ejecting her from group time with the children including the preparing and eating of family meals and while watching television, and ignoring her attempts to engage in family discussions. C.H.C. agreed that the distancing occurred but disputed this characterization, suggesting rather that he and the children were simply disengaging from C.C. in an effort to deescalate situations when she would become belligerent or belittling towards them.
- [14] Following an admissibility voir dire , a total of 17 surreptitiously obtained audio recordings made by C.C. of family conversations and arguments over the period March 28-July 7, 2022, were entered into evidence. C.H.C. testified that he was unaware the

recordings were being created, but agreed they evidenced the overall situation in the home accurately. In admitting these recordings I highlighted that the surreptitious recording of household conversations is an odious practice that should be discouraged ( Seddon v. Seddon , 1994 CanLII 3335 (BC SC) para 25), and that creating them raises questions of the parent's judgement ( TLA v. DAA , 2004 BCSC 1610 (CanLII) at para 31). I also noted that the use of this evidence can cut both ways, as while it may disclose some matters concerning the parenting skills or approach of the targeted party, the fact that the other party would resort to surreptitious recordings of private conversations may also be considered to determine their fitness to parent ( A.M.E.R. v. P.J.R. , 2003 BCSC 1466 (CanLII) at para 5).

## C.C.

- [15] I therefore begin by finding that C.C.'s making of the surreptitious recordings constituted a breach of faith in the privacy of communications with the children and C.H.C. This breach itself is a form of family violence, as it evidences an aspect of coercive and controlling behaviour directed at everyone in the family and has undoubtedly resulted in harm to both children's psychological and emotional security and well-being.
- [16] Within the earliest recording, the children initially tell C.C. they have 'no problem hanging out with her,' and tell her that the way she is reacting is itself stressful, which is neither acknowledged nor acted upon by C.C. When she challenged them, the children agreed that upon first being told their parents would separate they wanted to move between their mother's and father's home weekly.  They then changed their minds when they perceived C.C. had gotten 'greedy' by trying to 'extort $2,000 a month from dad.' They then told C.C. that they had asked C.H.C. for more information, but then suggested that C.H.C. 'never told us much' and 'barely told us anything.'
- [17] T.G.C.C. later told C.C. that shared parenting would have been a waste of time and that he and his sister had decided on their own, having seen who C.C. really was, they decided to live full-time with C.H.C. Within the judicial interview conducted with both adolescent children, they denied ever speaking in a substantive way with either

parent regarding conditions of the separation and divorce, and stated that any details they had learned was through eavesdropping on their parent's arguments.

- [18] C.C. also accused the children of lying to her within the first recording. She chastised them for holding their views and not agreeing with her, and told them they should be ashamed of themselves. She then made serious and sensitive allegations of infidelity against C.H.C. and strongly suggested to the children that C.H.C. had manipulated them. From the tone of P.S.C.C. and T.G.C.C., it is evident that they believed C.H.C. was innocent of all allegations and agree that they have been distancing themselves from C.C., but attributed this to her actions and behaviours and not C.H.C.'s.
- [19] In another conversation recorded in mid-May of 2022, C.C. detailed in an authoritarian manner that C.H.C. and both children refused to allow her to express her views. By the tone of his voice and words, T.G.C.C. was put on his guard and strongly rejected her suggestion. In a separate recording made later in May 2022, C.C. tried to initiate conversations with the children and when she was not answered, demanded they speak to her, even despite requests by the children to desist. Her tone and tenor, and the plain words she used, communicated her belief that the family was being unfair to her without a reasonable cause.
- [20] In yet another conversation T.G.C.C. told C.C. that his sister P.S.C.C. had played a larger role in raising him than C.C. C.C. responded that she has only recently stepped back to give the family more space and at the direction of C.H.C. At another point C.H.C. is heard telling C.C. that (recently) he had encouraged the adolescent children to spend time with her, that (the children) will spend time and stay in communication with her going forward, and that she will always be part of (the children's) lives.
- [21] The culminating incident that saw C.H.C. and the two adolescent children leave the family home in late August 2022 followed an emotionally charged verbal altercation among the whole family, part of which was videotaped by T.G.C.C. That video was also entered into evidence. Within that encounter C.C. insisted that C.H.C. inappropriately took personal items of hers, and then further alleged acts against the two children who

themselves speak in defence of their father and each other. C.C. accused C.H.C. of possessing pornography of a young girl who looked like P.S.C.C., stated that she had text messages in relation to his alleged infidelity, that C.H.C. had recently disposed of property he had improperly taken from his employer, and that he was being investigated for other 'serious' matters. Not captured in the video, a brief physical altercation subsequently ensued between C.H.C. and C.C. that resulted in her contacting police and alleging C.H.C. had assaulted her. C.H.C. and the children departed the residence later that same day and only returned once to gather additional personal items, and a subsequent police investigation was conducted but no charges were approved.

- [22] Observable within the video, and heard within many of the audio recordings, C.C. is heard speaking in a very distraught and agitated state, making serious accusations sometimes without obvious foundation, and rejected outright any view or attempt to disagree with her. Equally observable within the video, and many of the audio recordings, both adolescents speak to C.C. in a condescending and even hostile manner not uncommon to teenagers, and in the video T.G.C.C. can be seen making an obscene hand gesture that itself communicates moderate to extreme contempt directed at C.C.
- [23] I find that during the period following discussion of possible financial consequences of separation and a shared parenting model until C.H.C. and the children departed the home, C.C. herself engaged in further family violence. This family violence took the form of psychological and emotional abuse of P.S.C.C. and T.G.C.C. through C.C.'s heated discussions with them and refusal to consider their views. Likewise, she consistently refocused the toxic environment in the home onto herself, reflecting how it made her feel and how she was concerned, without reaching out to the children in any meaningful way. It also took the form of the accusations she made to the children regarding C.H.C., including allegations of infidelity and of viewing pornography.
- [24] This family violence is somewhat mitigated however by the children's own dismissive, contemptuous and hostile communications with C.C. C.C. was the adult in these situations, and no matter how inappropriate her children's communications or

tone was, it lay to her to remain 'the adult in the room,' and not to berate and belittle her young adults. It was inappropriate to make allegations of such an adult and sensitive nature to the children regarding their father, and I find this was an attempt to undermine his parenting role within the family.

[25] I do not however find that this family violence preceded this period, contrary to the allegations made by the children and as communicated to the court in their counsel's summary of their positions, as I will discuss further when canvassing the views of the children. I also do not find that this family violence occurred after C.H.C. and the children left the home. C.C. has subsequently only once had any personal contact with her daughter, at P.S.C.C.'s high school graduation that C.C. was not invited to but attended nevertheless as a member of the public. I do not find her simply approaching P.S.C.C. to congratulate her, and then departing, to be family violence - to the contrary, I find that C.C. engaged in no family violence following C.H.C. and the children departing the family residence.

[26] I also find that C.C. engaged in family violence through the allegations she made to the police in relation to C.H.C. during the period the parties lived separated but within the family home. These allegations of his infidelity with another workplace member, inappropriately using workplace resources to perform a background check on a friend of P.S.C.C.'s, and his alleged theft of workplace materials, were all investigated and no further action taken. These unsubstantiated allegations against C.H.C. were clearly spiteful and intended to interfere with his career, and possibly that of the other workplace member involved. Once C.H.C. and the children departed the family home, C.C. took part in no further family violence against C.H.C.

## C.H.C.

[27] C.H.C. testified that once he and C.C. had agreed to an ultimate separation and divorce, but prior to December 2021, they worked together in order to build a sufficient employment history for C.C. to qualify for a mortgage on her own. In January 2022, following C.C. receiving legal advice, she told C.H.C. of the possible financial consequences of the shared parenting they had been discussing, at which point he

testified that their conversations broke down. C.H.C. severed the couple's finances, telling the court this was necessary anyways, but he took this action at that time as he was anticipating a series of back-pay cheques and was concerned C.C. would remove funds inappropriately. I was provided no evidence of a basis to support this concern and I reject C.H.C.'s post-facto rationale based on C.C.'s unrelated actions as a valid basis for this belief.

- [28] C.H.C. set up a separate account and removed C.C. from all home bills including the mortgage, and conducted all financial dealings through his own personal account. Likewise, he began taking upon himself all home chores including grocery shopping and meal preparation, and discouraged C.C. from also participating. He testified this was to minimize conflict with C.C., which he attributed solely to her own actions and behaviours.
- [29] In March 2022, C.H.C. was confronted by C.C. upon her receiving a letter from a lawyer stating that the children had decided they wished to live exclusively with their father. Evidence from both C.H.C. and C.C. confirm that the atmosphere in the home to this point had been strained, and upon receipt of the letter, the home situation became toxic. C.C. was not made to feel welcome at family dinners or during other shared family time, again in C.H.C.'s view due to her own actions. In the surreptitious recordings in evidence, and from the testimony of both C.H.C. and C.C., everyone in the home contributed to this atmosphere although blame for the cause was disputed.
- [30] In another recording C.H.C. resist's C.C.'s stated desire to become more involved in the family home life, stating 'I'm hearing the poor me and how you want to be involved, and I'm also remembering the complaint that you filed against me at work, which I'm not allowed to speak about.' I find this evidenced his animus against C.C., and a complete lack of empathy for her wish to be involved within the family and with the children. In another recording C.H.C. assured C.C. he had not wanted the children to live with him full-time, but upon noting that the children had blocked her texts and calls, said he would not support shared parenting at that time due to the children's expressed desires.

[31] Subsequently in early May of 2022, C.H.C. stated he would never tell the children not to talk with their mom, have nothing to do with her, ignore her calls or exclude her from everything. While I have heard no evidence that C.H.C. ever did give these directions to P.S.C.C. or T.G.C.C., I do accept his evidence that he made zero effort to encourage them to contact their mother and in fact defended their 'right' to make such decisions.

[32] In a recording made in late May 2022, C.C. is heard trying to engage with T.G.C.C. and then speaking to C.H.C., confused, as in the past few days it appeared T.G.C.C.'s animosity towards her had begun to 'thaw' but then suddenly reversed and he was again refusing to speak to her. C.H.C. tells C.C., 'I prompted him to do that. Because I wanted him to try and get back intoâ€¦ but I prompted him to do that. I said go and see if your mom wants something.' Later in that recording C.H.C. tells C.C. that he has been encouraging the children to spend time with her and that they will spend time and stay in communication with her going forward, and that she will always be part of their lives.

[33] I find that throughout the period commencing when C.H.C. began severing C.C. from his accounts, until the August incident, he engaged in family violence by psychologically and emotionally abusing C.C. by placing unreasonable restrictions on C.C.'s financial autonomy within the family relationship. I heard no basis or evidence of past incidents to support these acts, aside from an unfounded allegation that C.C. could improperly take funds from their joint account. I also heard that these acts were done to encourage C.C. to move along with the process of separating. I reject both these arguments. The timing of these acts was no coincidence; it was done only after the potential financial consequences of a shared parenting arrangement were known.

[34] By excluding C.C. from the accounts and taking it upon himself to exclusively shop, cook and otherwise care for the household, C.H.C. actively denied C.C.'s agency as a parent within the home, a circumstance that was picked up upon by the children in criticizing her as was heard in the recordings. It was also evident that C.H.C. regularly

engaged in behaviour that had the effect of undermining C.C.'s authority or influence in the home and thus inserted a wedge between her and the children.

[35] Following C.H.C. and the children departing the residence, C.C. was removed as a school emergency contact, and all contact between her and the children was severed. C.H.C. would not share any school information or records with C.C. including report cards and attendance reports, which C.C. had to seek directly from the school. C.H.C. refused to share school photos that had been taken of the children, citing copyright law. C.H.C. agreed that he was aware of the court order to encourage the children to have a good relationship with C.C. and to make a real effort to maintain polite, respectful communications with her. Despite this, he testified that after moving out he did nothing to facilitate the children's contact with C.C., that he ignored C.C.'s requests to spend time with the children citing this was their desire, and that it would be negative to push something like that on the children.

[36] C.H.C. further chose to disregard that same court order and refused to advise C.C. of significant events affecting the children. He provided no updates on the children's education or health or on their achieving significant milestones typical of Canadian youth such as passing driving tests. C.H.C. even went so far as not to advise C.C. when P.S.C.C. had moved from his residence to relocate with a friend to Prince George, reasoning to the court that initially he was unsure if the move was permanent. Even when he became convinced the move was a 'significant event', a court hearing was required, as he didn't wish to act against P.S.C.C.'s desire not to tell her mother.

[37] Lastly, C.H.C. testified that despite a court order to enrol each of the children in counselling and to make all reasonable efforts to have them attend; his only effort was to make the appointment and then ask the children if they wished to attend. He told the court he thought encouraging the children to attend would be a negative experience and thus he did nothing further. At no time did C.H.C. ever seek to have the order varied, rather he simply chose not to obey it.

[38] I find that subsequent to the August 2022 incident, C.H.C. engaged in further family violence through ignoring or intentionally misinterpreting court ordered

requirements including encouraging contact and counselling for the children. C.H.C. is heard within a recording before moving out, stating that he didn't like counsellors and has 'always seen them as crackpots and snake oil salesmen,' a view that came across clearly during his testimony to the court.

[39] I find that asking the children once for their views, and then abandoning all parental influence to encourage them to attend counselling, is not taking reasonable efforts. It is instead evidence that C.H.C. dismissed counselling and permitted the children likewise, to avoid speaking to a third party trained to help youth transition from a broken home. This is also a form of family violence, as by selectively abandoning his parenting role when it impacted C.C.'s relationship with the children, he communicated his belief to the children that any relationship with her was not important.

[40] C.H.C. also unilaterally severed all exchange of information regarding the children to C.C., to the point of citing copyright law in regards to the school photos. This was a ridiculous rationale and I reject it entirely. On top of removing her from school contact documents and as a medical emergency contact, this evidenced a pattern of frequent, ongoing, and sustained psychological and emotional abuse of C.C. using the children that continues to this day. In the recordings he can be heard having encouraged the children to connect with their mother, and then withdrawing that encouragement, both of which they evidently responded to. I find that he continued using this pattern of passive controlling behaviour over the children, encouraging their independence when it was directed at C.C., thus signalling to them that a relationship with her was unimportant. In so doing I find that he has caused significant and ongoing harm to the children's psychological and emotional well-being.

## C. Views of the Adolescent Children

[41] A final consent order was made on September 20, 2022, by the Honourable Judge Wright, in part directing that CYLC appointed counsel would summarize the children's views and wishes pursuant to ss. 202 and 203 of the FLA . Terms of that order included that counsel would receive copies of all reports, records, Motions and

Applications, and would make submissions to the court including the views and positions advanced on behalf of the children.

[42] Children have the legal right to be heard in all matters affecting them including custody and parenting, and courts should, wherever possible, ensure those views are heard before making such significant decisions ( G.(B.J.) v. G. (D.L.), 2010 YKSC 44 (CanLII) at paras 2, 6). Children capable of forming their own views have the right to express those views, which may then be given due weight in accordance with the child's age and maturity ( Stav v. Stav , 2012 BCCA 154 (CanLII) at para 67 citing FLA s.37(2)(b) and the Convention on the Rights of the Child , Can. T.S. 1992 No. 3, Art. 12). Summarized hearsay evidence of children is intended to keep the child from becoming a witness in the litigation ( Goldsmith v Holden, 2020 BCSC 1501 (CanLII) at para 9) and counsel are appointed to protect the child's interests, including shielding them from negative parental conflict and the adversarial nature of family law proceedings ( D.M. v. C.R. , 2021 BCPC 318 (CanLII), para 22).

[43] Counsel with the CYLC had met with each of the children well over 10 times since October of 2022, and provided several summaries of their views to the court. Within those summaries both P.S.C.C. and T.G.C.C. alleged a general pattern of nonparenting by C.C., commencing when they were very young, that she essentially failed to parent them in any significant way and they had to 'learn how to take care of themselves,' leaving 'P.S.C.C. â€¦ [acting as] more of a parent.' C.H.C. was asked about this characterization, and testified this description was itself 'a bit of a stretch' and 'preposterous' given the children's ages and their levels of activity that would require parental support.

- [44] The adolescent children's summaries went on to variously describe C.C. as manipulative, blunt, controlling and a mean person, and that C.C. would engage in traumatizing public and private acts of discipline. In the summary of P.S.C.C.'s views she described being sprayed down in the family shower once by C.C. as an eight-yearold while fully clothed, just to wash conditioner from her hair. P.S.C.C. further alleged C.C. called her a 'bitch,' 'whore,' and 'slut' when C.C. did not approve of clothing she

was wearing, and that C.C. threatened to smash her phone with a hammer during an argument.

[45] P.S.C.C. also described C.C. as controlling and disrespectful of her boundaries, that C.C. raised her to become 'a housewife,' and that P.S.C.C. was subjected to years of yelling, mocking, and being called insulting names and body-shaming while also being required to wear makeup as early as 13 years of age. P.S.C.C.'s summary then stated that she gets along well with C.H.C., sharing activities including cooking, which struck me given her allegation that C.C. raised her to be 'housewife,' and that C.H.C. always made time for her despite his heavy work schedule.

[46] The CYLC summary of T.G.C.C.'s views indicate he also feels he did not ever have a good relationship with C.C., and he described her as 'manipulative,' 'controlling,' and a 'mean person,' who constantly started arguments with all family members. T.G.C.C. suggested that while growing up he did not feel cared for by his mother, recalling that from his childhood, she would sit around all day watching TV and that his interactions with her were more negative than positive. T.G.C.C. said that he and his sister learned to take care of themselves, and that P.S.C.C. did more parenting than C.C. Within T.G.C.C.'s summary, he described two public incidents where C.C. did 'horrible' things to him and P.S.C.C. that involved yelling at them and using physical force until they cried, and described her actions as irrational. In contrast, T.G.C.C. described his relationship with C.H.C. as good, despite occasional differences, they can talk through their conflicts and that C.H.C. is a good parent who shares activities with him.

[47] C.H.C. himself did not recall either of the public disciplinary incidents described by T.G.C.C. to the CYLC counsel and likewise testified that it was inconceivable that C.C. left the bulk of parenting duties to P.S.C.C. while he was out of the home. I find it striking that, had the incidents as described by T.G.C.C. occurred as summarized in the report, T.G.C.C. attributed no fault to C.H.C. for failing to act in defence of the children despite that he would have had to be present.

- [48] Throughout the surreptitiously obtained audio tapes, and in the video clip created by T.G.C.C., both P.S.C.C. and T.G.C.C. consistently, repeatedly, and strongly defend their father from any suggestion by C.C. that he is not helping her to parent. This sentiment is strongly echoed within the CYLC summary, where each child continues to attribute only positive traits to C.H.C., defending him as responding reasonably in all interactions with C.C., and attribute universally negative traits to C.C.
- [49] Having reviewed the CYLC summaries prepared for this court, I begin by noting that they were well prepared and counsel consistently strove within her professional bounds to represent the children's stated wishes. Children of sufficient maturity have a right to have their own wishes advanced by counsel and heard by the court, and counsel has a professional duty to ensure that these wishes are heard and that the children's rights are respected ( F.(M.) v. L.(J.) , 2002 CanLII 63106 (QC CA) at para 51). In speaking with their counsel, however, a significant contradiction was made apparent. Both P.S.C.C. and T.G.C.C. stated unequivocally they did not wish to discuss personal matters with a counsellor, as that person would be a stranger, and T.G.C.C. in particular felt it was unnecessary. This reported hesitation to disclose personal information to a stranger is directly at odds though with their demonstrated willingness to speak to counsel, herself a stranger, on multiple occasions. In these meetings, the children disclose a great deal of personal information, events that would have been difficult to relate, and their views of their own mother. Likewise, having never spoken to counsellors I am at a loss as to how T.G.C.C. could form the view that it is 'unnecessary.'
- [50] C.H.C. testified that he himself does not believe in counsellors, as in his experience having been referred to counsellor services through the course of his work, he has found them unhelpful 'snake oil salesmen' and a waste of time. I find it likely that this attitude was permitted to influence both P.S.C.C. and T.G.C.C.'s views towards counsellors, and was directly contrary to the orders of the court. I remind all that the direction was for the children to attend counselling - no direction was provided as to what that counselling was to work at achieving.

[51] Both P.S.C.C. and T.G.C.C. have effectively lived without a mother since August 2022.  In their judicial interview, they stressed that they did not feel C.C. was capable of ending her 'bad behaviour,' which they described as a demeaning, controlling, manipulative, and dominating personality that rejected their views and wishes. Court ordered counselling was not directed towards encouraging or discouraging further contact between C.C. and the children, and it is very likely that counselling would have addressed the very concerns raised by P.S.C.C. and T.G.C.C.

[52] Their feelings of stress at the thought of experiencing the 'dominating personality and rejection of their views and wishes' attributed to C.C. would likely have been a focus of any counselling, but we will never know as C.H.C. took it upon himself not to follow court orders. In several recordings, C.C. noted T.G.C.C.'s behaviour towards her would briefly become more positive; something that C.H.C. took credit for by 'prompting him to do that.' Obviously, even during the height of tensions within the family immediately prior to C.H.C. and the children moving out, he was able to exert that level of influence over T.G.C.C. That he steadfastly has refused to do so following their departure is noted, and will factor into my ultimate decision.

[53] It was not disputed that C.C. only provided the anticipated financial consequences of a shared parenting arrangement to C.H.C. in January of 2022. It follows then that the children must have concluded at some point thereafter that C.C. was 'greedy.' I note that contrary to what the children told me in their judicial interview, the plain meaning of their statements to C.C. in the first recording would indicate that they did, at the least, have some discussion with C.H.C. about the financial consequences of shared parenting. Knowledge of these details had to precede forming a belief that C.C. was 'greedy,' or that her intent was to 'extort $2,000 a month' from C.H.C.

[54] Given these conclusions of greed and extortion based on shared vs. exclusive parenting, they also had to have additional information against which to gauge the figure of $2,000 in order to form these beliefs - such a conclusion does arise in a vacuum. I find it unlikely that the adolescent children only learned these details through

eavesdropping on their parents arguments, much less formed such conclusions on their own. As pointed out by C.C.'s counsel, why would children care about financial issues that do not involve them, absent some sort of background or context? The shift in the atmosphere within the home occurred almost immediately after C.C. advised C.H.C. of the possible financial consequences. Given the only other source of such information and context was C.H.C.; I find it more likely than not that he did discuss the financial consequences of parenting models with the children.

[55] Thus, while I accept the stated views and wishes contained within the CYLC summary and as expressed to me during my judicial interview, I am left with significant concerns as to how these views were formed. Adolescents such as P.S.C.C. and T.G.C.C. are at a developmental stage of maturity and like all adolescents live under legal restrictions in recognition of this still developing maturity. This is publically acknowledged and given legal effect through restrictions such as driving restrictions and prohibitions on the consumption of alcohol and nicotine. I am mindful of this, while also acknowledging that the children are free to form their own views, thus any question of whether their views have been 'tainted' shall be 'left to be determined as a question of due weight to be given to the views' ( S.K. v. D.G. , 2022 ABQB 425 at para 179).

[56] On the whole, and having considered the allegations made by both children, I find that their memories and views have been tainted by the actions and inactions of C.H.C., and thus I give them reduced weight. I do not find that the actions alleged against C.C. amount to abuse or neglect, rather they are demonstrations of a rigid and unyielding parenting style. Likewise, I find that this parenting style does not warrant the fear and rejection displayed by both children.

## D. Estrangement vs. Alienation

[57] In family law, alienation has been found where a child rejects a parent despite little or no objectively reasonable cause. It is particularly apt where that rejection is the result of the other parent's hostility or antipathy towards the targeted parent combined with intentional or unintentional undermining of the parent - child's relationship. This can be the result of direct or indirect conduct, and where that conduct was expressly

designed to alienate a child from the other parent courts have found it to be both pernicious and unpardonable.

[58] Realistic estrangement on the other hand exists where a child understandably refuses contact with a parent due to that parent's behaviour, which may include physical or emotional abuse, rigid or restrictive parenting, immature and self-centred behaviour, and/or dysfunctional conduct arising from the parent's own psychological or psychiatric issues. There, a child's rejection of the parent may well be a reasonable and adaptive response to the estranged parent's behaviour ( N.R.G. v. G.R.G ., 2015 BCSC 1062 (CanLII) at paras 277, 278).

[59] A voir dire was held in this matter to determine the admissibility and potential use of a report prepared by Dr. Barbara Jo Fidler, Ph.D., Registered Psychologist, with regards to the phenomena she termed Parent Child Conflict Problems ('PCCP'), an umbrella term that includes parental alienation and estrangement. Relying upon the decision of White Burgess Langille Inman v. Abbott and Haliburton Co ., 2015 SCC 23 (CanLII), I found the proposed report was logically relevant and necessary to assist the court in drawing appropriate inferences with regards to PCCP, that no exclusionary rule precluded this court considering this evidence and that Dr. Fidler is well qualified to provide this evidence. Further, I weighed the benefits of admitting this evidence against any potential harm that could flow from admission. The report and viva voce evidence of Dr. Fidler were permitted for the purpose of identifying general signs of PCCP to include alienation and realistic estrangement and the research that assists in recognizing differences between these phenomena with regards to children aged prepubescent to age 18, as well as intervention outcomes for children within these age ranges.

[60] At a point following Dr. Fidler's evidence, C.C. testified to the court that she had personally communicated with the doctor, something Dr. Fidler expressly testified had not occurred. With the assistance of counsel it was determined that in fact C.C. was mistaken and had previously spoken to another counsellor. I was invited to draw the negative inference against C.C. that she was intentionally trying to mislead the court, however I decline to make such a finding. I find that C.C. was confused and was not

attempting to intentionally mislead the court. I also note, however, that this confusion evidenced the firmness and intransigence in her own beliefs heard in the audio recordings, and as described by the children in their CYLC summaries, and that it remains as one of her dominant personality traits.

[61] I accept Dr. Fidler's evidence that the multi-factor model to explain PCCP is preferable over the single-factor model, which almost takes a digital or 'either-or' approach by focusing upon either the favoured parent's, or the rejected parent's, conduct to explain the children's views against the rejected parent. This false dichotomy would suggest that a child is either a victim of abuse from the rejected parent, or is a victim of alienating behaviours by the favoured parent, but not both, and in effect denies the child's own agency in the situation. Dr. Fidler notes that it is not uncommon for parents to express negative sentiments of the other parent to their children before, during, and after separation, both overtly and covertly, and that these parental alienating behaviours are commonly both intentional and unintentional. Despite the wide prevalence of such parental behaviours, Dr. Fidler's evidence is that most children do not become alienated from a parent, which underscores the existence of multiple causes and contributing factors that have varying levels of influence and are present in a normal family dynamic.

[62] I accept Dr. Fidler's research that identifies that a myriad of dynamic and interacting factors exist, which can predispose, precipitate, and perpetuate children resisting or refusing contact with a parent, which include:

1. Child factors including age, cognitive capacity, temperament, social, emotional and academic adjustment pre and postseparation, vulnerabilities, special needs such as learning or physical disabilities, and coping skills/resilience. Present in this situation, I find that both adolescent children demonstrate a maturity, temperament, and cognitive capacity that is generally age appropriate and normal, having experienced low parental conflict prior to separation but higher conflict during the parental separation. No special needs for learning or physical disabilities are present, although coping skills and resilience are called into question given their poorly substantiated allegations of C.C.'s

parenting behaviour and their use of that as a basis for their complete rejection of her.

2. Parental conflict before and after the separation. In this situation direct parental conflict existed prior to separation and most markedly in the eight months leading up to C.H.C. and the children departing the family home. There are no allegations that intimate partner violence or any form of sexual abuse was present, and I reject the children's characterization of C.C.'s actions in their youth as anything beyond otherwise normal parental discipline. Neither parent used physical violence with either of the children, likewise neither parent exhibited dysfunctional conduct due to mental health. Following separation the parental conflict continued in the form of C.H.C.'s rejection of any contact with C.C., and his actions to devalue any attempt by her to remain within the lives of either of the children.
3. Sibling influence and relationships . I find the two siblings appear very close, speaking of the other in unfailingly positive terms to the point of attributing a lead-parenting role to P.S.C.C., and T.G.C.C. stepping into disputes between C.C. and P.S.C.C. to defend his sister. Their close relationship appears to reinforce each of their views in regards to both their parents, as is apparent in the CYLC summaries in evidence.
4. Favoured parent and rejected parent factors including parenting style and capacity, negative beliefs and behaviours, mental health, personality, responsiveness and willingness to change. I have heard no evidence that either C.H.C. or C.C. suffer from any form of mental health condition or are part of a concerning belief system. From the recordings in evidence, the testimony of both C.H.C. and C.C., and C.C.'s behaviour in regards to Dr. Fidler's evidence, it is evident that C.C. demonstrates a rigid parenting style that is self-centered on her belief that she knows best and includes one-way communication with the children, with minimal acceptance of the children's contrary opinions or views. Within the recordings, I also heard evidence of her use of emotional language to attempt to shame the children. C.H.C. would appear to largely utilize an authoritative parenting style, responding more openly to the children's views and then using reasoning and logical analysis to set limits or express his contrary views. These positive behaviours are contrasted with his demonstrated unwillingness to follow court orders made for the benefit of the children and his negative and largely unfounded views of C.C. that were likely communicated to the children, both intentionally and

unintentionally, and which C.H.C. has stubbornly refused to acknowledge or change.

5. Adversarial process and litigation . Both children formed an opinion in early 2022 regarding the perceived fairness of potential financial consequences of their parents separation as heard in the recordings, and do not appear to have waivered. It is likely that the ongoing nature of these proceedings have contributed to these adolescent children's behaviours and intransigence towards any form of contact with C.C.
6. Third parties such as aligned professionals including therapists, child protection workers, emergency room physicians, healthcare workers and lawyers, and extended family such as grandparents, new partners, neighbours and friends . Here, neither C.H.C. nor either of the children have accepted the assistance of therapists or counsellors and I have not heard any evidence of influence by any other person in their lives. As already noted they accepted the court ordered services of legal counsel and have communicated frequently with her, and may have incorrectly perceived her duty to communicate their views and stated wishes to the court as support for or vindication of those views and wishes.
7. Lack of functional co-parenting, and poor or conflictual parental communication . This factor is strongly present in this situation, as C.H.C. has demonstrated a complete unwillingness to co-parent by ignoring or refusing all requests for information or parenting time by C.C. Likewise, C.H.C. has consistently disregarded court orders to encourage a positive relationship between the children and their mother, or to attend and participate in counselling. I reject C.H C.'s reasoning that he did so to preserve his own relationship with the children and to not alienate them, or that he believes it would be unhelpful to 'force' or 'push' the children to take up any relationship with their mother, or that he was defending the children's right to choose contact with C.C. This justification is self-serving and not borne out in the evidence I do accept, which includes instances captured in the recordings where C.H.C. admits to influencing the children to spend time with C.C. prior to the August 2022 incident.
- [63] I likewise found helpful Dr. Fidler's research that identified commonly observed behavioural indications of alienation, from which I noted the following were engaged by evidence in this matter:

1. Weak, frivolous or absurd rationalizations for the rejection . Both children cite in the CYLC summary incidents that would have occurred years earlier as demonstrations of their mother's rejected behaviour. While the evidence I heard in the tapes does support the children's views that C.C. has a controlling parenting style and that, particularly in the final months she occasionally behaved irrationally, the utter rejection of any contact, including cards or letters, and digital communications on the phone, is extreme and out of proportion. I reject the suggestion that C.C. ever significantly, either verbally or physically, threatened the children. While both adolescent children allege in the CYLC summary that C.C. inappropriately employed corporal punishment and parenting behaviours while they were young children, these cited incidents do not justify the complete denial of contact that now exists. Likewise, both of these youth's rejection of counselling makes little sense, as they have freely shared intimate memories and views with legal counsel appointed by the court, and with the court itself, in apparent contradiction to their stated reluctance to share private information with strangers. I find their utter rejection of C.C. to be significantly out of proportion to the issues they claim exist or the behaviours they claim to wish to avoid.
2. Lack of ambivalence towards both parents; one is idealized without recognition of any negatives, while the other is devalued without acknowledgement of any positives . In the recordings, the children viewed C.C. in March 2022 as 'greedy' for trying to 'extort' $2,000 a month from C.H.C. as part of the separation, and in subsequent recordings, they would reflexively defend C.H.C. from any suggestion of wrongdoing within the marital relationship. These views do not appear to have changed, as C.H.C. is portrayed in the CYLC summaries in universally positive terms, while C.C. is only portrayed in negative terms without any positive acknowledgement.
3. Reflexive support of the favoured parent in all of the parental conflict. As heard in the recordings, both children demonstrated a significant animosity directed only at C.C., with little or no similar views expressed in relation to C.H.C. This animosity is then further reflected in the CYLC summaries wherein both children make almost universally negative comments and cite negative interactions with C.C., yet are equally universal in their positive views of C.H.C. This was most sharply drawn to the courts attention when during hearings early in this process P.S.C.C. cut her hand and required medical attention. It was conveyed to the court that had C.C. simply attended the hospital to see after her and ensure

her wellbeing, P.S.C.C. would 'fight her' and refused even that degree of contact.

4. Absence of guilt over cruelty to, or exploitation of, the rejected parent. Both children likewise demonstrated a complete lack of empathy for C.C. and the emotional toll that separation and divorce involves, and appear to have no insight into the emotional loss any parent would feel when all contact with their child is severed.

[64] Here, both P.S.C.C. and T.G.C.C.'s rejection of C.C. is plainly made out in the evidence, not only through C.H.C. and C.C.'s testimony but also through the CYLC report. Both P.S.C.C. and T.G.C.C. clearly hold generally positive views of C.H.C. while displaying, at best ambivalence, but most consistently a complete devaluation of C.C.

[65] Adopting Dr. Fidler's characterization of the multi-factor model to explain PCCPs present in this situation, and having made the findings I outlined above, I find that P.S.C.C. and T.G.C.C.'s behaviours are created from not only estranging actions by C.C., but also of alienating actions by C.H.C. C.C.'s behaviour, actions and conduct that culminated in the August 2022 incident set the conditions for both of her then adolescent children to understandably withdraw from her. This included her behaviour and actions directed at the children, but also her actions and allegations directed at C.H.C.

[66] I also find that the actions of C.H.C., beginning with the severing of financial ties and exclusion of C.C. from home activities, exacerbated P.S.C.C.'s and T.G.C.C.'s withdrawal of affections towards their mother. It was no coincidence that these actions commenced once the possible financial consequences of shared parenting became known. I find instead that this financial issue became the impetus for C.H.C.'s actions. It was striking that both of these adolescent children opined that the suggested support payments were 'extortion' by C.C. - child support is the right of the child, not of the receiving guardian. Child support in shared parenting uses federal guidelines and is based on the income generating ability of the separated guardians, and is intended simply to ensure that children of a broken relationship continue to enjoy a lifestyle with each parent that is not disproportionately reduced due to the breakdown in the family relationship.

- [67] C.H.C.'s estrangement from C.C. within their home was further compounded by her own allegations against him in the workplace. I do not go so far as to find he intentionally shared this with the children, however I do find that he took little effort to hide or mitigate his own feelings against C.C. in their presence and made zero effort to encourage a relationship between the children and C.C. By C.H.C.'s own words in the May recording he demonstrated that he could encourage positive contact by the children towards their mother if he so wished. I therefore reject entirely C.H.C.'s suggestion that he refrained from encouraging P.S.C.C. and T.G.C.C. from having any contact with their mother out of fear they would turn against him. These intentional and unintentional actions by C.H.C. occurred not only during the months leading up to the August 2022 incident, but have continued unabated ever since.
- [68] In Falvai v. Falvai , 2008 BCCA 503 (CanLII) at para 42, Madam Justice Smith for the court wrote, 'It is the responsibility of the custodial parent to facilitate meaningful access to the non-custodial parent that is in the best interests of the child'. This has not occurred here, and I find that both P.S.C.C. and T.G.C.C. are now significantly alienated from C.C. This alienation was precipitated somewhat by C.C.'s own temperament and inability to see her own faults, but exacerbated to a disproportionate level by C.H.C.'s ongoing attitude and behaviours. He intentionally severed all contact between C.C. and her children at their school and with their medical providers.
- [69] By adopting a self-professed passive attitude towards the children having any contact from their mother, at the same time as he actively rejected any contact with her himself, he reinforced a negative perception in the children that was out of proportion to what was reasonable in the circumstances. The adolescent children themselves were encouraged to conclude that C.C. was being greedy in trying to extort child support from C.H.C., and they were given almost no parental direction by C.H.C. to disabuse themselves of this belief.
- [70] Finding the best interests of the children is the function of this court, after hearing all the evidence including evidence of the children's wishes and their relationships with each of the parents. Where improper manipulation or alienation of the child is found, the

court will then weigh the children's wishes in the light of any evidence of that manipulation ( F.(M.) v. L.(J.) at para 52). Here I do find that P.S.C.C. and T.G.C.C.'s stated wishes and views, while communicated accurately by their CYLC legal counsel, are based on a distorted view of past events. Further, I find that while some estrangement exists, their views are more the product of alienating behaviours and the associated enmeshment with C.H.C. and his views. As a result and due to C.H.C.'s manipulations, the children's views will be given less weight.

## E. Family Bridges Counselling and Parenting Time

[71] C.C. asks this court to order that the Family Bridges, or Building Family Bridges, model of reintegration be followed. This is a peer reviewed program and is recognized within academic literature, following the main study by Dr. Warshak, Reclaiming ParentChild Relationships: Outcomes of Family Bridges with Alienated Children , Journal of Divorce &amp; Remarriage, 2019 Vol. 608, 642 - 647 ( M.S.R. v. D.M.R., 2022 BCSC 1398 at para 327). Upon a court finding, that extreme alienation has occurred, and where deemed in best interests of the child, this program would include court orders forcing 'restorative contact' with the rejected parent combined with a 'protective separation' of no contact from the favoured parent, for a period lasting at least 90 days.

[72] This parenting transition is combined with intensive residential counselling and therapy followed by out-patient counselling, all focused on overcoming the child's resistance to the rejected parent while rehabilitating any alienating behaviours of the favoured parent ( M.S.R. at para 330). By ordering this program, the forced participation of the child must be contemplated, including enforced transportation to the residential treatment site, lockdown while at that facility and then enforcement of no contact with the favoured parent during the 90-day protective separation.

[73] In researching the use of the Family Bridges model within BC, it is apparent that this is both a relatively new model and one that is directed when essentially all else has failed. In A.A. v. S.N.A ., 2007 BCCA 363 a ten-year-old child was found to have been alienated from her father due to the mother's highly manipulative and intransigent behaviour that was blind to the child's interests. Despite finding this, the trial judge did

not disturb the existing parenting arrangements. Relying in part on expert evidence, which included extensive therapy with the child, the unanimous court found that by focusing on the potential short term difficulties created by forcing a change in custody to the father, rather than giving the child's long-term interests paramountcy, the trial judge committed an error in law ( A.A. at para 27). Noting the efforts the father had taken to address the insecurity and anxiety anticipated with a change in custody, the trial decision was overturned and the child was ordered into the father's custody along with supervision and counselling ( A.A. , paras 28 - 31).

[74] In M.S.R. , the Court dealt with competing allegations of abuse made against the mother, and alienation alleged against the father. The 13 Â½ year old child had formed strong feelings against the mother following the couples separation, alleging the mother previously used abusive practices including limiting computer screen time, enforced chores, acting emotional in the child's presence and attending the child's school unannounced ( M.S.R. at para 293). That court had the benefit of a comprehensive s.211 report and held that the incidents as a whole did not amount to abuse or neglect warranting the child's displays of fear and rejection ( M.S.R. at para 294). The court found that the child was rebelling against the mother's rigid, rules based parenting style, while the father effectively 'abdicat[ed] his responsibilities as a parent' by promoting avoidance as a behaviour pattern. The court concluded this was 'tantamount to ongoing psychological abuse' ( M.S.R. at paras 315 - 319).

[75] In C.J.J. v. A.J ., 2016 BCSC 676, the court heard allegations that the father had alienated the couple's oldest 16-year-old child against the mother (paras 379 - 381). There, the court had extensive expert evidence including reports based on interviews of the child by several Doctors, psychologists, and clinicians, many noting the child's belief that the mother's behaviours caused the estrangement. The court considered this expert evidence and ultimately found severe alienation by the father. The court was then urged to consider the child's wishes, noting that the child was doing well in school and described as intelligent, mature and had firmly held views against residential treatment (paras 403 - 404). Noting that s. 37(2) of the FLA requires those views be considered, the court held this was not an ordinary case of a recalcitrant teenager and

gave reduced weight to the teen's wishes, and ordered residential reunification program to treat both the teen child and the father (paras 396 - 413).

[76] In M. v. M., 2015 BCSC 1297, the court likewise dealt with disputed allegations of alienation against the mother. The father felt that a course of therapist treatment, short of the extreme deconditioning characterized in the Family Bridges program, would be best (paras 204 - 211). Again with the benefit of extensive expert evidence including interviews of the children, the court found that alienation had not occurred and therefore declined to order a Family Bridges style of treatment in favour of a court directed outpatient form of reunification or rehabilitative therapy (at paras 316 - 317).

[77] My decision must consider the 'best interests of the children only' and consider all of the child's needs and circumstances including ten enumerated factors set out at s. 37(2) of the FLA . Any order must protect, to the greatest extent possible, the child's physical, psychological and emotional safety, security and well-being. Further, I may consider a person's conduct only if it substantially affects a factor set out in subsection (2), and then only to the extent it affects that factor ( FLA s.37(1) - (4).

[78] The long-term effects of parental alienation can be quite severe. The literature shows that the negative effects can include problems with self-esteem and identity ('if one parent is all bad then I must be part bad'), depression (present in 70% of alienation victims), problems with alcohol and drugs (present in 33% of alienation victims), and as adults they are more likely to have children who become alienated ( C.J.J. v A.J ., 2016 BCSC 676 at para 270).

[79] Having found family violence was committed in the past by C.C. and her estranging behaviour, the consistent and ongoing alienation perpetuated by C.H.C. is also form of psychological abuse ( C.J.J. at para 410). Given this family violence, I must take this into account within my order.

[80] As already noted, P.S.C.C. is no longer a child of the family and thus I will confine myself to considering only T.G.C.C.'s needs and circumstances when making this determination. While C.C. urges that I order T.G.C.C. participate in a Family

Bridges style of treatment, unlike all of the cases I have just cited I have no expert evidence involving actual interviews with, or input from, the children. Likewise, I was given only a vague outline of what the Family Bridges treatment sought would be comprised of, without actual facilities or therapists identified for me to consider in such an order. I do find that the status quo is not in T.G.C.C.'s best long-term interests, despite any short-term challenges that may be created. I also note that both P.S.C.C. and T.G.C.C. willingly accepted the courts direction when it came to speaking with court appointed legal counsel and with the court itself, and thus am confident they will continue to do so.

[81] Noting this lack of evidence combined with the understanding that T.G.C.C. remains in school, is doing well, and his education would understandably be interrupted, should I order the extensive Family Bridges level of treatment sought by C.C., I decline to do so at this time. I will, however, be making a comprehensive order directing T.G.C.C. to resume contact, and in time resume living with his mother under a shared parenting schedule and for both of them, and C.H.C., to take part in counselling. This plan will provide the option for P.S.C.C., at her own discretion, to also participate. I will also direct that this matter shall return to me in court no later than June 2025 for an update on the progress of the treatment plan I will set out, at which time I will consider further submissions on the Family Bridges treatment plan, and support, if necessary.

## F. Imputation of Income and Spousal Support

[82] Counsel for both C.H.C. and C.C. have suggested that this court should impute income to the other for the purposes of both child and spousal support. Arrears are also claimed, complicated by the fact that P.S.C.C. left C.H.C.'s home in mid-September 2023, to reside in another city. It is disputed when this became a 'significant event', and thus when child support requirements for her ended. In a hearing on the matter, C.H.C. conceded that P.S.C.C. had relocated as of 11 December 2023. It is not contested that P.S.C.C. returned to his home in April of 2024, and as of her 19 th  birthday at the end of October 2024, she is no longer a child of the marriage for child support purposes. C.H.C. also agreed that while living away from his home P.S.C.C. supported herself from her savings.

- [83] In an interim order regarding this matter I ordered C.H.C. to disclose all costs, supported by receipts, associated with P.S.C.C.'s relocation to support his position of when she had effectively left his home, but I received no such disclosure. As he again has not followed court direction, I am making a negative inference and find that P.S.C.C. left the family home, and was no longer a child of the marriage for the purposes of child support, for the period of October 1, 2023 until April 30, 2024. By virtue of her birthday in late October, child support for P.S.C.C. ends as of November 1, 2024.
- [84] Pursuant to the order of the honourable Judge Wright made on September 20, 2022, C.H.C.'s annual income was imputed to be $123,000 and C.C.'s income was imputed to be $56,000. On an interim and without prejudice basis, noting that the children lived primarily with their father, monthly spousal support of $1,199 to C.C. and child support of $869 to C.H.C. was ordered. This resulted in an offset taxable payment of $330 monthly to C.C. I understand this amount has been paid up until the date of this judgment.
- [85] C.H.C. continues to work in his position and is now again anticipating a promotion and further back pay. He was on paid administrative leave for almost four months related to the allegations made by C.C. after the August 2022 incident. He was also placed on paid medical leave from March 2023 - April 2024 due to what he testified was personal and workplace stress related to interpersonal conflict with a supervisor. While on these periods of leave, he occupied himself by woodworking, building furniture for his own use, and caring for those of his adolescent children living in the home. He testified that during those periods he could have sought other employment if permission was granted by his workplace. He did not seek such permission, and disputes any imputation of income over these periods. In 2023, his employer reached an agreement with the employees and as a result, his income will again be adjusted upwards 4% each year from his 2022 income for the years 2023 and 2024. Notwithstanding this, C.H.C. urges this court to simply rely on the income as stated in his 2023 Notice of Assessment, $129,903, but concedes the income for 2024 should be set at $148,366 to reflect this raise.

[86] From C.C.'s Notice of Assessment ('NOA') for 2022 she actually earned $51,263, having worked full-time with her former employer, and part-time for Core Practice. This reflects the additional employment she took on when the couple decided to separate, as her NOA income for 2021 was $39,956. She testified that she left her former employer once C.H.C. returned to work at the end of 2022 due to the stress of him being in the workplace, and that she was let go from Core Practice due to declining hours and conflicts with her current job with her current employer. She began working with her current employer in February 2023 at .8 of a full time position and testified this was in accordance with medical advice. A single medical letter was provided into evidence from 2023 that stated simply that C.C. 'should continue to work 4 days per week due to [sic] medical concern.'

[87] Counsel for C.H.C. urged this court to adopt the reasoning in Davidson v. Patten , 2021 ONCJ 437 at para 15 in rejecting C.C.'s evidence of a medical condition. The court in Davidson held that when a party resists a claim for imputation of income for medical reasons this must be supported by evidence confirming at least the diagnosis, prognosis, treatment plan and level of compliance, and specific information connecting the medical condition to the ability to work. In that case, the applicant provided only a single letter stating he was unable to work for one month due to a medical condition.

[88] I noted that C.H.C. himself only provided a very heavily redacted medical note, dated April 18, 2023, which simply stated C.H.C. was unable to work beginning March 15, 2023. Although, I am uncertain whether the current state of the Canadian medical services can routinely support the requirements set out in Davidson , in the absence of any other evidence I note that neither C.H.C. nor C.C. have provided sufficient evidence of a medical condition that prevented them from working to their full capacity.

[89] I was provided no evidence, nor was any suggestion made, of any employment C.H.C. could have taken up over the periods he was away from work, and I will not speculate as to potential employment or self-employed income that could have occurred. Recognizing that both parties have provided insufficient evidence upon which imputation of income may be resisted, and as permitted at s.19(1) of the Federal Child

Support Guidelines (SOR/97-175) and the reasoning set out in Donovan v Donovan , 2000 MBCA 80 (CanLII) at para 21, I therefore will attribute full time employment to C.C. but beginning as of May 2024. This recognizes that over the same period as C.H.C., she no doubt suffered as much or even greater mental stress due to the separation, hers amplified by the effective loss of her children.

- [90] As a result, I determine that C.C. had employment income in 2022 of $51,263, in 2023, she earned $52,848 less union dues of $1,094 as reflected in her tax returns, and for 2024, I impute an income of $54,587 based on her hourly wages, imputed if working full time from May - December, less $1,130 union dues. This will result in child support due from C.C. to C.H.C. of:
- a.  2022: September - December $799 per month, total $3,196 in 2022;
- b.  2023: January - August $806 per month; $6,448; September December $486 per month; $1,944, total of $8,392 in 2023;
- c.  2024: January - April, November - December, $502 per month; $3,012; May - October, $831 per month, $4,986, total of $7,998 for 2024; and
- d.  2025: Beginning in January, $502 per month, total of $502 for 2025.
- [91] The total amount of child support due from C.C. as of the date of this decision therefore is $20,088.
- [92] Turning to the question of spousal support, I find that C.H.C.'s employment income for 2022 to be $129,903; I impute his income for 2023 to be $135,099 less $1,865 union dues, and for 2024 $148,366 less $1,906 union dues. Section 160 of the FLA imposes a duty upon a spouse to pay spousal support having consideration for the objectives of spousal support found at s.161, and then in accordance with the factors set out at s. 162. The objectives of spousal support are
- a.  To recognize any economic advantages or disadvantages to the spouses arising from the relationship between the spouses or the breakdown of that relationship;

- b.  To apportion between the spouses any financial consequences arising from the care of their child, beyond the duty to provide support for the child;
- c.  To relieve any economic hardship of the spouses arising from the breakdown of the relationship between the spouses; and
- d.  As far as practicable, to promote the economic self-sufficiency of each spouse within a reasonable period of time.

[93] Further, in determining C.H.C.'s duty to provide support for C.C., the amount and duration of spousal support is set considering the conditions, means, needs, and other circumstances of each spouse. This includes the length of time the spouses lived together, functions performed by each spouse during this period and any agreement or order relating to either spouse's support.

[94] C.H.C. argues that this court should await his final income from 2023 before making any order for support. He further argues that C.C. should receive spousal support at the lower end due to her own behaviours as the couple separated and after, and the existence of available job opportunities in the area. He also argues C.C. has failed to maximize the income generating possibilities of funds the couple divided upon the sale of common property as she has kept them in a savings account rather than investing them more aggressively. Lastly, he suggests that she has spent unwisely as evidenced by her maintaining a storage locker while living in rental accommodations. Counsel suggests that C.C. also failed to provide evidence of her employment prior to the relationship, and thus this court cannot accurately determine what her earning potential would have been, and now suggests that C.H.C. provided equal care for the children prior to separation.

[95] I find I have sufficient evidence upon which to impute incomes, and will do so now. I completely reject the suggestion that I should consider C.C.'s failure to provide evidence of her earning potential from 2004 and earlier. The couple had been together since 1999, lived common law in 2002, and married in 2004. It was uncontroverted that they moved for C.H.C.'s employment and advancement, which was successful - he has been promoted once and I was advised he anticipates another promotion.

[96] It is not disputed that C.C. took on part-time employment once the children were both in school, even periodically travelling to the lower mainland to maintain her security clearance required to continue working at her former employment. That employment apparently is no longer possible, given that C.H.C. continues his career and there is no other local employment location for C.C. to continue hers.  She continues to be economically disadvantaged by C.H.C.'s employment.

[97] Likewise, C.H.C. himself testified that particularly when the children were younger C.C. provided the bulk of childcare and that it was 'preposterous' to suggest otherwise, given the children's ages and his own work schedule. I therefore reject his suggestion now, through counsel, that it was shared childcare. C.C. sacrificed her career opportunities in order to raise the couple's children. Then with the support of C.H.C., she took on part-time employment along with her former full time employment in 2021 in order to qualify for a mortgage on her own. Her subsequent underemployment as I have found will be accounted for.

[98] I also reject the suggestion that C.C. should have better invested the residue of funds she received from the division of family assets, or is unnecessarily maintaining a storage locker. Separation of the couple did not imbue her with the financial acumen of Warren Buffet. I heard no evidence, nor even suggestion, that she has engaged in an extravagant lifestyle or otherwise consumed her resources irresponsibly. I therefore reject these arguments.

[99] C.C. argues that spousal support should be set at the high end, noting that she sacrificed a career out of the home for the vast majority of their time as a couple with children, permitting C.H.C. to progress his career. C.C. also submits that she has suffered negative health effects due to the effective loss of her children combined with the obvious animus of C.H.C., who would not have even told her that P.S.C.C. had moved to another city had she not discovered this for herself. As a woman in the latter part of her employable years, she will suffer permanent impairment of employment opportunities. C.C. further suggests that had C.H.C. acted to encourage the children to maintain contact, spousal support within the middle range would likely have been

appropriate, as she would then have likely been better able to maintain full-time employment and have enjoyed the comfort of her own children, as C.H.C. has.

[100] I accept that C.C. has, like many who remain home to provide the majority of childcare for dependent children, suffered an impairment to her employment opportunities. This is exacerbated by her age, as she will have a shortened window within which to build a career and see advancement. I also accept that she has suffered negative health effects due to the manner of the separation and the effective loss of her children during a very formative part of their lives. As I have already spoken to, she bears some responsibility for how this began, but has had no opportunity to repair that damage through the intransigence, inaction and spiteful indifference of C.H.C. Within his own evidence, C.H.C. suggested taking a significant amount of time on paid leave from his duties, due in part to the stress of the separation but also due to workplace stress and difficulties. If he found stress interfered with his ability to work, while still enjoying the love and support of their children, then it is no stretch to find C.C. suffered equally or greater.

[101] C.C.'s conduct during the separation was far from blameless. She made allegations against C.H.C. that had to be investigated and on one occasion, saw him arrested, however, by his own evidence, no further action was taken and fortunately, he suffered no career consequence as a result. She testified that she has attended counselling since the separation, however her own intransigence and inability to accept that she may be wrong and others correct is still very much a part of her, as shown in regards to the evidence of Dr. Fidler.

[102] Thus, I find that having lived in a marriage-like relationship for almost 20 years, C.H.C.'s employment situation benefited with C.C. predominately maintaining their home and children. C.C. is further disadvantaged by entering the workforce on a fulltime basis at a later point in her employable life, and no extraordinary financial consequences arise due to care of the children. C.C. has suffered additional economic hardship due to the actions and purposeful inactions of C.H.C. with regards to those children. Lastly, I find that C.C. has taken concrete steps to promote her own economic

self-sufficiency, however will be permanently economically impaired as I have spoken to.

[103] I am therefore setting spousal support at the medium range, which I find applied commencing upon the couple's separation. As of today's date spousal support is set at $2,166 monthly, and I am further directing that spousal support shall continue to be payable at the medium range until 2040, subject to annual review. When this matter returns in court to me for the update on the family reunification process, in addition to hearing further submissions if necessary on the need for a Family Bridges program, I will also consider whether this should have an impact on the level of spousal support ordered.

[104] In looking at the arrears of child and spousal support payable, spousal support as set out in the Spousal Support Advisory Guidelines for those months the couple were separated in 2022 is $1,493 monthly for a total of $5,972; in 2023 spousal support payable while P.S.C.C. was living with C.H.C. was $1,601 monthly and while she was living away from C.H.C. was $1,887 for a total of $20,356; in 2024 spousal support payable while P.S.C.C. was living with C.H.C. was $1,823 monthly and while she was living away from C.H.C. and after her 19 th  birthday was $2,166, for a total of $24,620. Thus total spousal support payable as of this decision is $50,948. Deducting child support payable from C.C. to C.H.C., $20,088, and accounting for the $330 monthly support offset that was paid by C.H.C. to C.C. over this same period totalling $9,240, support arrears from C.H.C. to C.C. will be set at $21,620 as of today's date.

## FINAL ORDER

[105]  A1: Upon the court being advised that the name and birth date of each child is as follows: P. S. C.C., born [omitted for publication] and T.G.C.C., born [omitted for publication], and that as of the date of this decision T.G.C.C is the sole child subject to the terms of this order.

[106]  B1: The court is satisfied that C.C. and C.H.C. are the guardians of the children under s.39(1) of the Family Law Act .

[107] B6: Pursuant to s. 40(3)(a) of the Family Law Act , C.C. will solely have all of the parental responsibilities for the child as set out in s. 41 of the Family Law Act .

[108] B11: C.C. and C.H.C. shall have the obligation to advise each other of any matters of a significant nature affecting the child:

- i. C.C. shall immediately arrange for a non-residential counselling and therapy Plan to be developed, under the overall guidance of a single Counsellor or therapist, for the Guardians and T.G.C.C., with the option for P.S.C.C. to participate at her own discretion;
- ii. C.C. has full discretion in retaining the Counsellor responsible to develop the Plan, which shall be communicated to C.H.C. and T.G.C.C. immediately upon confirmation. C.C. shall pay the initial retainer as required and provide C.H.C. a receipt, and C.H.C. shall repay half the retainer within 30 days;
- iii. The Plan shall be focused on family reintegration therapy to include parent coaching and counselling, parent and co-parent education, and any other therapeutic interventions to assist the family to remedy their current difficulties, address parenting and coparenting, improve strained parent-child relationship(s), and otherwise assist in implementing this court ordered reunification;
- iv. If required, the Counsellor may communicate with any other professionals deemed necessary to implement the Plan, and the Guardians will execute any required authorizations to permit this exchange of information between those professionals;
- v. The Guardians and T.G.C.C., and P.S.C.C. at her own discretion, shall forthwith engage and fully participate in the Plan which is designed on their behalf including completion of and return of any intake questionnaire within one week of receipt;
- vi. The Guardians and T.G.C.C., and P.S.C.C. at her own discretion, shall attend counselling and therapy as and when directed by the Plan;
- vii. The Plan shall be as intensive as possible and should include as much therapy and counselling as the Counsellor considers necessary, taking into account T.G.C.C.'s school requirements, and with accommodation as required in the event P.S.C.C. choses to participate;

- viii. The Counsellor may make such recommendations to the Guardians, counsel and the Court as required to ensure sufficient information has been obtained, including information regarding transition and transfer protocols, location, behaviour and pacing of parent-child contact within a determined parenting time schedule, parent communication and child-related information sharing protocols;
- ix. The Counsellor may, at their sole discretion, share information with other family members participating in the Plan, and provide status or summary process reports and recommendations to the Court and other relevant professionals;
- x. In the event T.G.C.C. attends post secondary education, necessary accommodations for his attendance via digital or remote means of communication shall be enacted as directed by the Counsellor, until T.G.C.C.'s 19 th  birthday;
- xi. The Guardians shall adjust their respective work and personal schedules to fully participate in the Plan;
- xii. The Guardians shall comply with all directions and treatment recommendations made by the Plan, including any rules of engagement as directed by the Counsellor;
- xiii. In order to ensure compliance with the Plan, there shall be no vacations or trips, shopping or otherwise, outside of the [omitted for publication] region of British Columbia up to and including the commencement of school in the fall of 2025 by T.G.C.C. unless expressly authorized by the Counsellor;
- xiv. No other mental health professional not previously approved of by the Counsellor, unless required for C.H.C. in direct relation to his employment only, shall be permitted to become involved with either of the Guardians or T.G.C.C. during the execution of the Plan without the approval of the Counsellor;
- xv. If the Counsellor is of the view that T.G.C.C. requires in-person therapy C.C. shall make all necessary arrangements, and T.G.C.C. shall attend, during any school breaks, in order to ensure the recommended therapy takes place;
- xvi. Without constraining the Counsellor, the Plan should, absent extraordinary circumstances, which shall be set out in writing by the

Counsellor, see C.C. having supervised parenting time with T.G.C.C. within two weeks of today's date and thereafter unsupervised parenting time in accordance with the Plan. Unsupervised parenting time shall include overnight access as directed by the Counsellor. The supervised parenting time should involve other members of C.C. or C.H.C.'s family, if considered in the best interests of T.G.C.C.;

- xvii. There shall be no recording of any participant while engaged in the Plan unless previously agreed to by the Counsellor and all parties;
- xviii. The Counsellor shall direct the Guardians regarding communications with the children when they are under the parental care of the other Guardian, and the Guardians shall adhere to the Counsellor's recommendations in that regard;
- xix. C.H.C. shall immediately and on an ongoing basis communicate clearly to the children that he does not oppose and, indeed, that he supports parenting time with C.C.;
- xx. The Guardians shall take all reasonable steps to encourage the children's relationship with the other party and to influence all third parties to do likewise;
- xxi. The Guardians shall not make or convey any negative or disparaging comments about one another to the children or in their presence, and the Guardians shall not directly or indirectly make disparaging remarks about the other to third parties;
- xxii. The Counsellor is directed to prepare and file with the court a report, due by June 1st, 2025, outlining:
- a.  The compliance of the Guardians and T.G.C.C. with the Plan to date;
- b.  Recommendations pertaining to the feasibility of and timetable for a shared parenting plan between the Guardians and T.G.C.C.;
- c.  Recommendations pertaining to the re-establishment of contact between C.C. and P.S.C.C., if possible;
- d.  The Guardians' participation and progress in the Plan to that date including whether progress is being achieved and, if not, the reasons therefore;

- e.  Outlining future parenting and treatment recommendations considered appropriate for the future protection and promotion of T.G.C.C.'s psychoemotional well-being and development; and
- f. Any other recommendations or observations the Counsellor may wish to bring to the court's attention, including but not limited to whether the parties are genuinely engaging in the Plan and its objectives.
- xxiii. The costs of the participation in the Plan and the preparation of the Counsellor's report shall be shared equally by the Guardians;
- xxiv. A 2 hour hearing shall be set by the JCM, to take place following receipt of the Counsellor report and no later than 30 June 2025, for the purpose of confirming adherence with the Plan as I have outlined, and to consider whether further court orders including a Family Bridges Program is required in the best interests of the child;
- xxv. In the event there is a dispute regarding implementation of the Plan by either Guardian, an; application to review the Plan shall be made and the matter set into court on 7 days notice to the other parties. I will remain seized of this matter for the next 2 years.
- [109] B68: C.H.C. is found to be a resident of British Columbia and is imputed to have a guideline annual income of $ 146,460.
- [110] B74: C.H.C. shall pay to C.C. as spousal support the sum of $2,166 per month, commencing on February 1st 2025, and continuing on the first day of each and every month thereafter. Either party may apply pursuant to s. 168 to review the spousal support payable after July 31st 2025, and spousal support will continue to be payable until further order of this court or July 31 2040. A review will be permitted on the following grounds: ongoing hardship and its effect on economic ability of the parties; and the following matters will be considered for the purposes of a review: Adherence with the Plan as developed and overseen by the Counsellor, and the degree of family reconciliation achieved between both guardians and the child.
- [111] B56: C.C. is found to be a resident of British Columbia and is imputed to have a guideline annual income of $ 54,587.

[112] B61: C.C. shall pay to C.H.C. the sum of $502 per month for the support of the child, commencing on February 1 2025, and continuing on the first day of each and every month thereafter, for as long as the child is eligible for support under the Family Law Act or until further court order.

- [113] B77: The arrears of support payable by C.H.C. to C.C. are fixed as of today's date in the amount of $ 21,620 inclusive of statutory interest if any.
- [114] D53: For as long as the child is eligible to receive child support, the parties shall exchange: (a) copies of their respective income tax returns for the previous year, including all attachments, not later than July 1 st  each year; and (b) copies of any Notice of Assessment or Reassessment provided to them by the Canada Revenue Agency, immediately upon receipt.

[115] This concludes my order.



The Honourable Judge D. Reeves Provincial Court of British Columbia

"""

inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=max_seq_length).to("cuda")
outputs = model.generate(**inputs, max_new_tokens=256, temperature=0.7)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
````

1.  **Separate System Instructions and Case Content**: I will extract the detailed system instructions and schema from the large `prompt` variable and store them separately. The actual legal case content will also be isolated.
2.  **Chunk the Case Content**: The legal case content will be divided into smaller, overlapping chunks. Each chunk will be sized to fit within the language model's context window when combined with the system instructions.
3.  **Process Chunks (Map Phase)**: For each chunk, I will construct a prompt that includes the system instructions and a specific chunk of the case. The model will analyze this chunk and generate a partial legal reasoning trace and/or extract relevant information. These partial outputs will be collected.
4.  **Aggregate Results (Reduce Phase)**: After processing all chunks, a final call to the language model will be made. This final prompt will include all the partial analyses from the chunks and instruct the model to synthesize them into a single, comprehensive legal reasoning trace and a structured JSON output, ensuring it adheres to the full schema provided.

This approach will allow us to analyze the entire document while respecting the memory limitations of the GPU and the context window of the language model.

**Reasoning**:
Now that the system instructions, schema, and legal case content are separated, I will proceed to chunk the legal case content using the tokenizer to ensure token-based splitting and overlap, considering the model's maximum sequence length.



````python
import textwrap

# Define the maximum sequence length and reserve tokens for output
# max_seq_length is already defined as 2048
output_token_reserve = 500
max_input_tokens = max_seq_length - output_token_reserve

# A simple character-based chunking strategy if a tokenizer is not readily available for token counting
# For accurate token counting, an actual tokenizer should be used. However, given the context,
# a character-based approximation is suitable for this step to demonstrate chunking.
# Let's approximate token count by characters (e.g., 1 token ~ 4 characters for English)
# We'll use a slightly more conservative character count to stay within limits.
chars_per_token = 4 # Average character per token approximation
chunk_size_chars = max_input_tokens * chars_per_token
overlap_chars = int(0.15 * chunk_size_chars) # 15% overlap

# Function to split text into overlapping chunks
def split_text_into_chunks(text, chunk_size, overlap_size):
    chunks = []
    start_index = 0
    while start_index < len(text):
        end_index = min(start_index + chunk_size, len(text))
        chunks.append(text[start_index:end_index])
        if end_index == len(text): # Reached the end of the text
            break
        start_index += (chunk_size - overlap_size)
        # Ensure we don't start a new chunk beyond the text length if overlap is too large
        if start_index >= len(text):
            break
    return chunks

# Split the legal case content into overlapping chunks
legal_case_chunks = split_text_into_chunks(legal_case_content, chunk_size_chars, overlap_chars)

print(f"Legal case content split into {len(legal_case_chunks)} chunks.")
print(f"First chunk (partial): {legal_case_chunks[0][:500]}...")
print(f"Last chunk (partial): {legal_case_chunks[-1][-500:]}...")
````
