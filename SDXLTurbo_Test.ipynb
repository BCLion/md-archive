```python
# CELL 1: INSTALL & LOAD SDXL-TURBO
!pip install diffusers accelerate torch transformers huggingface_hub

from diffusers import StableDiffusionXLPipeline
import torch

print("[EYE] Loading SDXL-Turbo...")
# "Turbo" is the optimized/faster version
pipe = StableDiffusionXLPipeline.from_pretrained(
    "stabilityai/sdxl-turbo",
    torch_dtype=torch.float16,
    variant="fp16", # Optimized memory
    use_safetensors=True
)
pipe.to("cuda")

print("[EYE] SDXL-Turbo Ready.")
```

```python
# CELL 2: GENERATE ARTIFACT LIBRARY ## ORIGINAL
from pathlib import Path

output_dir = Path("temple_artifacts")
output_dir.mkdir(exist_ok=True)

artifacts = [
    "A glowing neon obelisk, cyberpunk style, red and blue circuitry",
    "A holographic data key, glowing gold wireframe",
    "Ancient stone wall with embedded glowing circuit board"
]

print("[EYE] Generating Library...")
for i, prompt in enumerate(artifacts):
    print(f"Generating Artifact {i+1}: {prompt}")

    image = pipe(
        prompt,
        guidance_scale=7.5,
        num_inference_steps=30,
        width=512, height=512 # Square textures work best for Roblox
    ).images[0]

    # Save to Colab
    image.save(f"{output_dir}/artifact_{i+1}.png")

print("[EYE] Library Generated. Check folder icon on left.")
```

```python
# CELL 2: GENERATE ARTIFACT LIBRARY (Metallic Edition)

from pathlib import Path

output_dir = Path("temple_artifacts")
output_dir.mkdir(exist_ok=True)

# THE CLEAN NEGATIVE PROMPT
NEG_PROMPT = "glass bottle, medicine, drugs, alcohol, whiskey, liquid splashing, text, watermark, blurry, multiple items, collage, borders"

# YOUR UPDATED METALLIC ARTIFACTS
artifacts = [
    "A glowing neon obelisk, cyberpunk style, red and blue circuitry, dark background, centered, isolated",
    "A holographic data key, glowing gold wireframe, dark background, centered, isolated",
    "Ancient stone wall with embedded glowing circuit board, dark background, centered, isolated",
    "A jagged glowing neon healing crystal, cyberpunk aesthetic, red and blue glowing circuitry, floating in a dark void, centered",
    "A decorative Victorian-style metallic canteen, ornate brass and copper filigree, a bright glowing turquoise light emitting from the seams, steampunk aesthetic, isolated on a solid dark background, centered, 8k resolution, cinematic lighting, magical healing artifact.",
    "An ornate Victorian ornate gold canister, silver filigree, glowing spectral green liquid, eerie gothic haunted mansion aesthetic, isolated, dark background"
]

# Short semantic names for the filenames
short_names = [
    "cyber_obelisk", "data_key", "circuit_wall",
    "healing_crystal", "healing_canteen", "gold_canister"
]

print("[EYE] Forging Metallic Foundation Library...")
for i, prompt in enumerate(artifacts):
    name = short_names[i]
    print(f"Generating ({i+1}/{len(artifacts)}): {name}")

    image = pipe(
        prompt=prompt,
        negative_prompt=NEG_PROMPT,
        guidance_scale=2.5,     # Slightly higher for metallic detail
        num_inference_steps=4,  # Turbo speed
        width=512, height=512
    ).images[0]

    # Save to Colab
    image.save(f"{output_dir}/{name}.png")

print(f"\n[EYE] Forging Complete. Download your assets from the '{output_dir}' folder!")
```
